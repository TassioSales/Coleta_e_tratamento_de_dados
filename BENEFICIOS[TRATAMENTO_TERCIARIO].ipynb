```python
# Importação de Bibliotecas
import glob # O glob é uma biblioteca para coletar e manipular o endereço de pastas no computador
import os
import pandas as pd
import numpy as np
```

```python
# Variáveis globais.
caminho_padrao = "/content/drive/MyDrive/Projetos_Tassio_Sales/beneficios_ao_cidadao-12-2021/dados_tratamento_secundario/"
caminho_salvar = "/content/drive/MyDrive/Projetos_Tassio_Sales/beneficios_ao_cidadao-12-2021/dados_tratamento_terciario/"
```

```python
def listarArquivos(path):
  # Listando arquivos contidos na pasta.
  caminho = path + "/*.txt"
  arquivos = glob.glob(caminho)
  print("Arquivos Contidos: {}".format(arquivos))
  return arquivos
```

```python
# Definindo pasta para salvar os dados concatenados.
file_name = caminho_salvar + "beneficios_ao_cidadao.txt"

# Pasta para ser acessada.
print("Acessou a pasta: {}\n".format(caminho_padrao))

arquivos = listarArquivos(caminho_padrao)
print()

# Criar arquivo .txt.
with open(file_name, 'w') as outfile: # Criando arquivo vazio e abrindo este arquivo.
    outfile.write("NOME_BENEFICIARIO;CPF_BENEFICIARIO;NIS_BENEFICIARIO;BENEFICIO\n") # Escreve a primeira linha do arquivo como cabeçalho.
    for fname in arquivos: # Abre arquivo por arquivo do diretório.
        print("### Lendo arquivo: {}".format(fname))
        with open(fname, encoding='latin-1') as infile: # Realiza a leitura do arquivo.
            next(infile) # Pula a primeira linha do arquivo que é o cabeçalho de colunas.
            # Para cada linha dentro do arquivo, escreve a linha no arquivo final.
            for line in infile:
                outfile.write(line) # Escreve a linha no arquivo final.
        infile.close() # Fecha o arquivo dos dados.
outfile.close() # Fecha o arquivo final.

print()  
      
print("### Salvando arquivo: {}".format(file_name))
```
